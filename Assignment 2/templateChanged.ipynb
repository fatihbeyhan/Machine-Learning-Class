### b)  Visualizing the data:

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

def plot_image(row):
    image = row.values.reshape(28, 28)
    plt.imshow(image, cmap = matplotlib.cm.binary,
    interpolation='gaussian')
    plt.axis("off")
    plt.show()
```
